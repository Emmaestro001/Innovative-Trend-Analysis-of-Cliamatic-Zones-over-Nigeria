```python
#import the necessary packages

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

Import the different spreadsheets containing the parameters for each location
cal - calabar,
ibn - ibadan,
kuk - kukawa,
kan - kano.

```python
cal=pd.read_excel('climate corr.xlsx', sheet_name ='calabar', index_col='Year')
cal.head()
```

```python
kuk=pd.read_excel('climate corr.xlsx', sheet_name ='kukawa', index_col='Year')
kuk.head()
```

```python
kan=pd.read_excel('climate corr.xlsx', sheet_name ='kano', index_col='Year')
kan.head()
```

```python
ibn=pd.read_excel('climate corr.xlsx', sheet_name ='ibadan', index_col='Year')
ibn.head()
```

FIND THE CORRELATION BETWEEN PARAMETERS FOR EACH LOCATION

```python
ibn_corr = ibn.corr()
ibn_corr
```

```python
cal_corr = cal.corr()
cal_corr
```

```python
kuk_corr = kuk.corr()
kuk_corr
```

```python
kan_corr = kan.corr()
kan_corr
```

PLOT THE HEAT MAP SHOWING THE RESULTS FROM THE CORRELATION TABLES

```python
sns.heatmap(cal_corr, annot=True)
plt.title('(a) CALABAR - Tropical Monsoon', size=15)
plt.savefig('CAL_CORR.png', dpi = 300, bbox_inches='tight')
```

```python
sns.heatmap(ibn_corr, annot=True)
plt.title('(b) IBADAN - Tropical Savannah', size=15)
plt.savefig('IBN_CORR.png', dpi = 300, bbox_inches='tight')
```

```python
sns.heatmap(kuk_corr, annot=True)
plt.title('(d) KUKAWA - Warm Desert', size=15)
plt.savefig('KUK_CORR.png', dpi = 300, bbox_inches='tight')
```

```python
sns.heatmap(kan_corr, annot=True)
plt.title('(c) KANO - Warm Sermarid', size=15)
plt.savefig('KAN_CORR.png', dpi = 300, bbox_inches='tight')
```
