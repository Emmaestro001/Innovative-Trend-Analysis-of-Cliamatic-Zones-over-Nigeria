THE COLUMNS HAVE BEEN SEPARATED BY INDICATING THE FIRST HALF (fh), AND SECOND HALF (sh) FOR EACH PARAMETER TO ENABLE US CARRY OUT THE INNOVATIVE TREND ANALYSIS 

```python
cal= pd.read_excel('climate research analyzed.xlsx', sheet_name ='cal')
cal.head()
```
