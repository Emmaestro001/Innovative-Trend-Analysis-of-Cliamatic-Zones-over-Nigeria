```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
T = pd.read_excel("Box.xlsx", 'T')
VP = pd.read_excel("Box.xlsx", 'VP')
REF = pd.read_excel("Box.xlsx", 'REF')
AP = pd.read_excel("Box.xlsx", 'AP')
R = pd.read_excel("Box.xlsx", 'R')
TL = pd.read_excel("Box.xlsx", 'TL')
EPT = pd.read_excel("Box.xlsx", 'EPT')
PT = pd.read_excel("Box.xlsx", 'PT')
```

```python
T.plot(kind='box', subplots=False, layout=(2, 2), figsize=(6,4), ylabel='Temperature (K)')
plt.savefig('CLI_T.png', dpi = 300)
```

```python
VP.plot(kind='box', subplots=False, layout=(2, 2), figsize=(6,4), ylabel='Vapour Pressure (hPa)')  
plt.savefig('CLI_VP.png', dpi = 300)
```

```python
AP.plot(kind='box', subplots=False, layout=(2, 2), figsize=(6,4), ylabel='Atmospheric Pressure (hPa)')  
plt.savefig('CLI_AP.png', dpi = 300)
```

```python
REF.plot(kind='box', subplots=False, layout=(2, 2), figsize=(6,4), ylabel='Refractivity (N-Units)') 
plt.savefig('CLI_REF.png', dpi = 300)
```

```python
R.plot(kind='box', subplots=False, layout=(2, 2), figsize=(6,4), ylabel='Mixing Ratio (g/kg)')  
plt.savefig('CLI_MR.png', dpi = 300)
```

```python
TL.plot(kind='box', subplots=False, layout=(2, 2), figsize=(6,4), ylabel='$T_L$ (K)')  
plt.savefig('CLI_TL.png', dpi = 300)
```

```python
EPT.plot(kind='box', subplots=False, layout=(2, 2), figsize=(6,4), ylabel=r"$\theta_T$ (K)") 
plt.savefig('CLI_EPT.png', dpi = 300)
```

```python
PT.plot(kind='box', subplots=False, layout=(2, 2), figsize=(6,4), ylabel=r"$\theta$ (K)")  
plt.savefig('CLI_PT.png', dpi = 300)
```
